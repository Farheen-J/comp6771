17. Visualize blurred image vs deblurred image

```python
# Load the images
blurred_image = mpimg.imread('../output/test_deblurred_images/blurred_image.jpg')
deblurred_image = mpimg.imread('../output/test_deblurred_images/deblurred_image.jpg')

# Display the images side by side
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(blurred_image)
ax[0].axis('on')
ax[0].set_title('Blurred Image', fontsize=12)
ax[1].imshow(deblurred_image)
ax[1].axis('on')
ax[1].set_title('Deblurred Image', fontsize=12)
plt.show()
```
